Shown below is a sample of the dataset, illustrating that the AI generated images have a companion real generated image with similar subjects and composition.

It can also be seen that label 1 is the AI-generated class and label 0 is the real image class.

```python
from PIL import Image
import matplotlib.pyplot as plt
for i in range(0,6,2):
    fig, axes = plt.subplots(1,2)
    for j in range(0, 2):
        im = Image.open(f'{file_path}{df_train['file_name'][i+j]}')
        axes[j].imshow(im)
        axes[j].title.set_text(f'{df_train['label'][i+j]}')
    plt.show()

```
